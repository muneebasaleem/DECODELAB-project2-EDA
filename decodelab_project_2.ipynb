**Project Overview**

This project focuses on performing Exploratory Data Analysis (EDA) on the ull_dataset.csv dataset. **Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Upload and Load Dataset**

```python
from google.colab import files

uploaded = files.upload()

df = pd.read_csv("Dataset for Data Analytics (1).csv")

df.head()
```

**Dataset Overview**

```python
print("Dataset Shape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nDataset Info:")
df.info()
```

**Dataset Information**

```python
df.info()
```

**Check Missing Values**

```python
missing_values = df.isnull().sum()

print("Missing Values in Each Column:")
print(missing_values)
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

**Descriptive Statistics**

```python
df.describe()
```

**Basic Statistics (Mean, Median, Count)**

```python
# Mean
print("Mean Quantity:", df["Quantity"].mean())
print("Mean Unit Price:", df["UnitPrice"].mean())
print("Mean Total Price:", df["TotalPrice"].mean())

# Median
print("\nMedian Quantity:", df["Quantity"].median())
print("Median Unit Price:", df["UnitPrice"].median())
print("Median Total Price:", df["TotalPrice"].median())

# Count
print("\nTotal Orders:", df["OrderID"].count())
```

# **Distribution Analysis**

**Quantity Distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Quantity"], kde=True)
plt.title("Quantity Distribution")
plt.show()
```

**Total Price Distribution**

```python
plt.figure(figsize=(8,5))
sns.histplot(df["TotalPrice"], kde=True)
plt.title("Total Price Distribution")
plt.show()
```

# **Trend Analysis**

**Monthly Sales Trend**

```python
df["Date"] = pd.to_datetime(df["Date"])

monthly_sales = df.groupby(
    df["Date"].dt.to_period("M")
)["TotalPrice"].sum()

monthly_sales.plot(
    kind="line",
    figsize=(10,5),
    marker="o"
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)

plt.show()
```

**Product Sales Trend**

```python
product_sales = df.groupby("Product")["TotalPrice"].sum()

product_sales.sort_values(ascending=False).plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")

plt.show()
```

**Payment Method Analysis**

```python
df['PaymentMethod'].value_counts()
df['PaymentMethod'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)

plt.title("Payment Methods")
plt.ylabel("")
plt.show()
```

**Referral Source Analysis**

```python
df['ReferralSource'].value_counts()
plt.figure(figsize=(7,4))

sns.countplot(
    data=df,
    x='ReferralSource'
)

plt.title("Referral Sources")
plt.xticks(rotation=45)
plt.show()
```

**Outlier Detection**

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["TotalPrice"])

plt.title("Outlier Detection in Total Price")

plt.show()
```

**Order Status Analysis**

```python
df['OrderStatus'].value_counts()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="OrderStatus"
)

plt.title("Order Status Distribution")
plt.xticks(rotation=45)

plt.show()
```

**Correlation Analysis**

```python
numeric_cols = df[
    ["Quantity",
     "UnitPrice",
     "ItemsInCart",
     "TotalPrice"]
]

plt.figure(figsize=(8,5))

sns.heatmap(
    numeric_cols.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

**Top 10 Highest Value Orders**

```python
df.nlargest(10, 'TotalPrice')[['OrderID', 'Product', 'TotalPrice']]
```

**Key Observations**

```python
print("1. Most common order status:", df['OrderStatus'].mode()[0])

print("2. Most used payment method:",
      df['PaymentMethod'].mode()[0])

print("3. Top referral source:",
      df['ReferralSource'].mode()[0])

print("4. Average order value:",
      round(df['TotalPrice'].mean(), 2))

print("5. Maximum order value:",
      round(df['TotalPrice'].max(), 2))
```
